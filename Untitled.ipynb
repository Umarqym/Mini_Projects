```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('PC.csv', encoding = "ISO-8859-1", engine='python')
```

```python
df.head(5)
```

Assessing all the null values in each column

```python
for i in df.columns:
    print(df[i].isna().mean())
```

```python
df.dtypes
```

Filling the null values for numeric columns with '0' and 'unknown' for string based columns

```python
df.iloc[:,16:22] = df.iloc[:,16:22].fillna(0)
```

```python
df.iloc[:,:16] = df.iloc[:,:16].fillna('Unknown')
```

```python
for i in df.columns:
    print(df[i].isna().mean())
```

Changing the numberic columns to int data type as the columns are related to people, which cannot be float or decimal numbers

```python
df.iloc[:, 16:22] = df.iloc[:, 16:22].astype(int)
```

```python
df.dtypes
```

Removing columns that are not required and creating a new column for surviviors

```python
df.drop(['Aircraft', 'Operator', 'Registration', 'Schedule','MSN', 'YOM',
       'Flight no.' ], axis = 1, inplace = True)
```

```python
df['Total on Board'] = df['Crew on board'] + df['Pax on board']
```

```python
df['Survived'] = df['Total on Board'] - df['Total fatalities']
```

```python
df.columns
```

Analysing total fatalities in relation to causes of plane crashes

```python
cause = df.groupby('Crash cause').sum().sort_values('Total fatalities', ascending = False)
```

```python
fig = plt.Figure(figsize=[15,8])
ax = fig.add_axes([0.1,0.1,0.9,0.9])
ax.bar(cause.index, cause['Total fatalities'])
ax.set_xlabel('Causes', fontsize=20)
ax.set_ylabel('Total Fatalities', fontsize=20)
ax.set_title('Plane Crash Fatalities', fontsize=20)
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)

ax.set_axisbelow(True)
ax.yaxis.grid(color='gray', linestyle='dashed', alpha=0.2)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```

```python
fig
```

Analysing data in terms of flight stage

```python
flight_stage = df.groupby('Flight phase').count().sort_values('Date', ascending = True)
flight_stage
```

```python
fig1 = plt.Figure(figsize=[12,8])
ax1 = fig1.add_axes([0.1,0.1,0.9,0.9])
ax1.barh(flight_stage.index, flight_stage['Date'])
ax1.set_xlabel('Number of Crashes', fontsize=20)
ax1.set_ylabel('Flight Stage', fontsize=20)
ax1.set_title('Plane Crash Flight Stage', fontsize=20)
ax1.tick_params(axis='x', labelsize=12)
ax1.tick_params(axis='y', labelsize=12)

ax1.set_axisbelow(True)
ax1.yaxis.grid(color='gray', linestyle='dashed', alpha=0.2)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
```

Analysing crashes interms of flight type

```python
flight_type = df.groupby('Flight type').sum().sort_values('Total fatalities', ascending = False).head(10)
flight_type
```

```python
fig2 = plt.figure(figsize=[15,8])
ax2 = fig2.add_axes([0.1,0.1,0.9,0.9])
ax2.bar(flight_type.index, flight_type['Total fatalities'])
plt.setp(ax2.get_xticklabels(), rotation=30, ha='right')
ax2.set_xlabel('Flight Type', fontsize=20)
ax2.set_ylabel('Number of Fatalities', fontsize=20)
ax2.set_title('Flight Type Fatalities', fontsize=20)
ax2.tick_params(axis='x', labelsize=12)
ax2.tick_params(axis='y', labelsize=12)

ax2.set_axisbelow(True)
ax2.yaxis.grid(color='gray', linestyle='dashed', alpha=0.2)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
```

```python
passenger_flights = df[df['Flight type'] == 'Scheduled Revenue Flight'].groupby('Crash cause').sum().sort_values('Total fatalities', ascending = False)
passenger_flights
```

```python
fig3 = plt.figure(figsize=[15,8])
ax3 = fig3.add_axes([0.1,0.1,0.9,0.9])
x = np.arange(len(passenger_flights.index))
w = 0.2
ax3.bar(x-0.4, passenger_flights['Crew fatalities'], w)
ax3.bar(x-0.2, passenger_flights['PAX fatalities'], w)
ax3.bar(x, passenger_flights['Other fatalities'], w)
ax3.bar(x+0.2, passenger_flights['Total fatalities'], w)
plt.xticks(x, passenger_flights.index)
plt.legend(['Crew fatalities', 'Passenger Fatalities', 'Other Fatalities', 'Total Fatalisties'], loc='upper right', ncol = 4)
ax3.set_xlabel('Crash Cause', fontsize=20)
ax3.set_ylabel('Casualties', fontsize=20)
ax3.set_title('Passenger Flight Casualties', fontsize=20)
ax3.tick_params(axis='x', labelsize=12)
ax3.tick_params(axis='y', labelsize=12)

ax3.set_axisbelow(True)
ax3.yaxis.grid(color='gray', linestyle='dashed', alpha=0.2)
ax3.spines['right'].set_visible(False)
ax3.spines['top'].set_visible(False)
```

Same Data presented on stacked bar chart

```python
fig4 = plt.Figure(figsize=[15,10])
ax4 = fig4.add_axes([0.1,0.1,0.9,0.9])

ax4.bar(passenger_flights.index, passenger_flights['Crew fatalities'], color = 'b')
ax4.bar(passenger_flights.index, passenger_flights['PAX fatalities'], bottom = passenger_flights['Crew fatalities'], color = 'orange')
ax4.bar(passenger_flights.index, passenger_flights['Other fatalities'], bottom = passenger_flights['Crew fatalities'] + passenger_flights['PAX fatalities'] , color = 'g')

ax4.legend(['Crew Fatalities', 'Passenger Fatalities', 'Other Fatalities'])
ax4.set_xlabel('Crash Cause', fontsize=20)
ax4.set_ylabel('Casualties', fontsize=20)
ax4.set_title('Passenger Flight Casualties', fontsize=20)
ax4.tick_params(axis='x', labelsize=12)
ax4.tick_params(axis='y', labelsize=12)

ax4.set_axisbelow(True)
ax4.yaxis.grid(color='gray', linestyle='dashed', alpha=0.2)
ax4.spines['right'].set_visible(False)
ax4.spines['top'].set_visible(False)
```
